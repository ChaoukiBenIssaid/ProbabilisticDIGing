### Logistic Regression

```python
X = np.load("datasets/reduced_covtype_features.npy")
y = np.load("datasets/reduced_covtype_target.npy")
n = 20
iter_max = 1200
learning_rate = 0.0005
```

```python
p = [1, 1, 1, 1, 1, 1, 10, 10, 10, 10, 10, 10, 20, 20, 20, 20,20,20, 7,7]
p = [50*i for i in p]
X_split = [X[sum(p[:k]):sum(p[:k]) +p[k], :] for k in range(n)]
y_split = [y[sum(p[:k]):sum(p[:k]) +p[k]] for k in range(n)]
train_test = [train_test_split(X_split[i], y_split[i]) for i in range(n)]
act_function = lambda x : 2*(x > 0) - 1
```

```python
dgd = sim_logistic_regression(train_test, n, iter_max, learning_rate, act_function, lmbd = 1, algo="dgd", standardize = True, seed = 10)
dig = sim_logistic_regression(train_test, n, iter_max, learning_rate, act_function, lmbd = 1, algo="dig", standardize = True, seed = 10)
```

```python
plt.plot(dgd["comm_costs"], dgd["accuracies"], label = "dgd")
plt.plot(dig["comm_costs"], dig["accuracies"], label = "dig")
plt.legend()
plt.xlabel("communication cost")
plt.ylabel("accuracy")
plt.savefig("letsgoooo.png")
```
